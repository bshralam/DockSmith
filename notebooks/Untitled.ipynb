```python
# Cell 1: Imports and ligand path
from IPython.display import display, HTML
import py3Dmol
import os

# Path to ligand file
ligand_file = '../results/best_poses/ligand_0_out.pdbqt'  # Change this to your actual ligand output file
receptor_file = '../data/mpro_clean.pdbqt'  # Change this to your actual receptor file

```

```python
# Cell 2: Function to read file and convert to string
def read_pdbqt(filename):
    with open(filename, 'r') as f:
        return f.read()
ligand_block = read_pdbqt(ligand_file)
receptor_block = read_pdbqt(receptor_file)
```

```python
# Cell 3: Visualize with py3Dmol
view = py3Dmol.view(width=600, height=400)
view.addModel(receptor_block, 'pdbqt')
view.setStyle({'model': -1}, {'cartoon': {'color': 'spectrum'}})
view.addModel(ligand_block, 'pdbqt')
view.setStyle({'model': 1}, {'stick': {'colorscheme': 'cyanCarbon'}})
view.zoomTo()
view.show()
```
